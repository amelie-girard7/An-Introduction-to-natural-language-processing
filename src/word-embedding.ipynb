```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import matplotlib.pyplot as plt

# Load images directly using plt.imread
apple_image = plt.imread('./img/apple.png')
orange_image = plt.imread('./img/orange.png')
cherry_image = plt.imread('./img/cherry.png')
mobile_phone_image = plt.imread('./img/iphone.png')
apple_logo_image = plt.imread('./img/apple-logo.png')

# Create a scatter plot without points, just to get the axes
plt.figure(figsize=(8, 5))
plt.scatter([], [])  # Empty scatter just to get the axes
plt.title('A toy example (vector space simplified to 2D)')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.xlim(0, 6)
plt.ylim(0, 6)

# Add grid to the plot
plt.grid(True, which='both', color='lightgray', linestyle='-', linewidth=0.5)

# Define zoom levels for each image, with apple at 50% of its original size
zoom_levels = {
    'apple': 0.05,  # 50% smaller zoom for the apple image
    'orange': 0.1,  # Current zoom level for the orange image
    'cherry_image': 0.05,
    'mobile_phone': 0.05,  # Current zoom level for the mobile phone image
    'apple_logo': 0.1,  # Current zoom level for the apple logo image
    'cherries_image': 0.05,
}

# Add images to plot as annotations with offset
for (x, y), image, zoom_level in zip(
        [(1, 1), (1, 2),(0.5,1), (5,5), (5,4)],
        [apple_image, orange_image,cherry_image, mobile_phone_image, apple_logo_image],
        zoom_levels.values()):
    image = OffsetImage(image, zoom=zoom_level)
    ab = AnnotationBbox(image, (x, y), frameon=False)
    plt.gca().add_artist(ab)

plt.show()
```
