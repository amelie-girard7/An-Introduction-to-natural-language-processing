```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Terms for demonstration
terms = ['the', 'petruchio', 'prospero', 'lear', 'macbeth']
# Corresponding tf-idf values for each term (arbitrarily chosen for demonstration)
tf_idf_values = [0, 633.05, 527.44, 847.55, 1058.53]

# The index for each term
indexes = np.arange(len(terms))

# Bar chart
bar_width = 0.35
opacity = 0.8

# Plotting each term with its tf-idf value
rects = plt.bar(indexes, tf_idf_values, bar_width,
                alpha=opacity,
                color='b',
                label='tf-idf')

# Labeling
plt.xlabel('Terms')
plt.ylabel('tf-idf Value')
plt.title('tf-idf values for different terms')
plt.xticks(indexes, terms)
plt.legend()

# Display the graph
plt.tight_layout()
plt.show()
```
